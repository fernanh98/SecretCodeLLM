## Game

```python
board = Board()

model = "mistral"

model = HumanLLMModel(
    model_name="Human",
)

fer = Captain(
    name="Fer", 
    model=model
)

sofi = Guesser(
    name="Sofi", 
    model=model
)

antonio = Captain(
    name="Antonio", 
    model=model
)

miriam = Guesser(
    name="Miriam",
    model=model
)

team_blue = Team(color="blue", players=[fer, sofi])
team_red = Team(color="red", players=[antonio, miriam])

secret_code_game = SecretCodeGame(
    board=board,
    team_blue=team_blue, 
    team_red=team_red
)
```

```python
secret_code_game.play()
print("Game is over\n")
print(secret_code_game.results)
```
